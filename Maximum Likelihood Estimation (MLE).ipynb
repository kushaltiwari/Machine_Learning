# Maximum Likelihood Estimation (MLE)

\begin{equation}
    D = \{x_{1}, x_{2}, \ldots x_{n}\}
\end{equation}

Each datapoint $x_{i}$ is collected from a sample of data that i.i.d. that are normally distributed and parameterized by two parameters mean and variance. $\mathbb{N} \sim (\mu, \sigma^{2})$

\begin{equation}
    \theta_{MLE} = \underset{\theta \in \Theta} {argmax} \prod_{i=1}^{n} \mathbb{P}(D | \theta)
\end{equation}

\begin{equation}
    \theta_{MLE} = \underset{\theta \in \Theta} {argmax} \prod_{i=1}^{n} \mathbb{P}(x_{i} | \theta)
\end{equation}

\begin{equation}
    \theta_{MLE} = \underset{\theta \in \Theta} {argmax} \prod_{i=1}^{n} \frac{1}{\sqrt(2\pi\sigma^{2})} exp-(\frac{1}{2\sigma^2} ({x_{i}-\theta})^2)
\end{equation}

The likelihood function can be written as

\begin{equation}
    \theta_{MLE} = \underset{\theta \in \Theta} {argmax} (\frac{1}{\sqrt(2\pi\sigma^{2})})^{n} exp-(\frac{1}{2\sigma^2} \sum_{i=1}^{n}({x_{i}-\theta})^2)
\end{equation}

Sweep $\theta$ through different sets of $\Theta$ and get value of $\theta$ that maximizes the value of $\mathbb{P}(D|\theta)$. $\theta$ is the mean of the ditribution. We compute $\mathbb{P}(D|\theta)$ for various mean i.e. $\theta$ and choose $\theta_{MLE}$ that maximizes the function. We are trying to find the mean ($\theta$) that best estimates the mean of the given distribution (D).


## Breaking down Likelihood function

$\sum_{i=1}^{n}({x_{i}-\theta})^2$ : Sum of squared deviation of each datapoint from the assumed $\theta$. $\theta$ is one of our assumed mean. This is an equation of variance. 

$exp-(\frac{1}{2\sigma^2} \sum_{i=1}^{n}({x_{i}-\theta})^2)$: Now we take the negative exponential of the variance. The negative exponential tends to zero when the variance is high. So negative exponential gives maximum value when the variation is minimum. Therefore, the $\theta$ value that corresponds to minimum variance is the "Maximum Likelihood Estimation" Or The best mean that represents the given distribution $x$.



```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
mean, std = 4.5, 2

# We take Normally distributed i.i.d. data for specified mean and standard deviation (x in the equation)
D = np.random.normal(mean, std, 10000)
plt.plot(D)
```

```python
# theta values to sweep across
sweep = np.arange(-10,10,0.5)


# Calculate Likelihood for every theta
#L = [np.exp(-np.sum((D - theta)**2)) for theta in sweep]


# Use "mean" instead of "sum" because negative exponential of large number becomes zero.
# Therefore instead of "sum" we use "mean" to keep from the exponential value becoming zero.
L =[]
check = 0
for theta in sweep:
    likelihood = np.exp(-np.mean((D - theta)**2)/(2*(std**2)))
    L.append(likelihood)
    # Store the value of theta that gives the maximum likelihood 
    if likelihood > check:
        val = theta
    check = likelihood
        
        
# Value of D (dataset) at which Likelihood function is maximum.
print('MLE:', val)


# Plot
plt.plot
plt.xlabel('D')
plt.ylabel('Probability')
plt.plot(sweep,L)
plt.plot((val, val),(max(L), 0))
```
